````python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials


client_id = env["client_id"]
client_secret = env["client_secret"]
client_credentials_manager = spotipy.oauth2.SpotifyClientCredentials(client_id, client_secret)
spotify = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
````

````python
from pathlib import Path
from urllib.request import urlretrieve


def download_mp3_from_sid(sid: str) -> bool:
    # print(sid)
    if Path(os.path.join(env["DATASET_PATH"], "spotify_sample",
                         sid[0], sid[1], sid[2], sid + ".mp3")).exists():
        return False
    try:
        res = spotify.track(sid)
    except Exception as e:
        print(e)
        return False
    url = res.get("preview_url", None)
    if url is not None:
        save_dir = os.path.join(
            env["DATASET_PATH"], "spotify_sample", sid[0], sid[1], sid[2])
        Path(save_dir).mkdir(exist_ok=True, parents=True)
        try:
            urlretrieve(url, os.path.join(save_dir, sid + ".mp3"))
        except Exception as e:
            print(e)
            return False
        return True
    else:
        return False
````
